```python
import csv
with open('data/hygfull.csv') as f:
    stars = list(csv.DictReader(f))
```

```python
import pylab as pl
import numpy as np
import math
tau = 2.0 * math.pi
```

```python
import spectral_classification
color_chart = spectral_classification.build_color_chart('starcolors.txt')
```

```python
def paren(spectral_class):
    letter = spectral_class[:2]
    roman = spectral_class[2:] or 'V'
    return '{}({})'.format(letter, roman)

print paren('G8III')
print paren('B9')
```

```python
stars2 = [
    star for star in stars
    if star['ColorIndex'] and
       paren(star['Spectrum']) in color_chart
    ]
print len(stars2), 'of', len(stars), 'have good colors'

star_magnitudes = np.array([ float(star['AbsMag']) for star in stars2 ])
star_colors = np.array([ float(star['ColorIndex']) for star in stars2 ])
star_spectra = np.array([ star['Spectrum'] for star in stars2 ])
```

```python
print star_magnitudes[:5]
print star_colors[:5]
print star_spectra[:5]
```

```python
hr_axes = [-0.5, 2, 15, -5]
```

```python
plot = plt.figure().add_subplot(111)
plot.scatter(star_colors, star_magnitudes, s=1,
             c=(0, 0, 0), linewidths=0)
plot.axis(hr_axes)
```

```python
star_rgbs = [
    color_chart[paren(spectral_class)]
    for spectral_class in star_spectra
    ]
```

```python
plot = plt.figure().add_subplot(111, axisbg='k')
plot.scatter(star_colors, star_magnitudes, s=1,
             c=star_rgbs, linewidths=0.0)
plot.axis(hr_axes)
```

```python
hr_range = [[-5, 15], [-0.5, 2]]
```

```python
heatmap, yedges, xedges = np.histogram2d(
    star_magnitudes, star_colors, bins=50, range=hr_range)
extent = [xedges[0], xedges[-1], yedges[-1], yedges[0]]
plot = plt.figure().add_subplot(111)
plot.imshow(heatmap, aspect=0.08, extent=extent)
```

What if we want some color?

```python
colorgrid = np.zeros((50, 50, 3))

ymin, ymax, xmin, xmax = -5, 15, -0.5, 2
y_range = ymax - ymin
x_range = xmax - xmin

for i in range(len(stars2)):
    xbin = 50.0 * (star_colors[i] - xmin) / x_range
    ybin = 50.0 * (star_magnitudes[i] - ymin) / y_range

    if not 0.0 <= xbin < 50.0:
        continue
    if not 0.0 <= ybin < 50.0:
        continue

    xbin = int(xbin)
    ybin = int(ybin)
    color = color_chart[paren(star_spectra[i])]
    colorgrid[ybin][xbin] += color

print heatmap[10][10]
print colorgrid[25][20:25]
colorgrid /= colorgrid.max()
print colorgrid[25][20:25]
```

```python
plot = plt.figure().add_subplot(111)
plot.imshow(colorgrid, aspect=0.08, extent=extent)
```
